```python
# creating the bigram frequency for the bible
biblebigrams = list(nltk.bigrams(biblewords))
print(biblebigrams[:50]) # printing hte first 50
```

```python
# creating the bigram frequency for the fellowship of the ring
fellowshipbigrams = list(nltk.bigrams(fellowshipwords))
print(fellowshipbigrams[:50]) #printing the first 50
```

```python
from nltk.collocations import * #importing the collocation finder module
```

```python
bigram_measures = nltk.collocations.BigramAssocMeasures() # define a variable for the bigram measures
biblefinder = BigramCollocationFinder.from_words(biblewords) # running the bigram finder on the bible
fellowshipfinder = BigramCollocationFinder.from_words(fellowshipwords) # running the bigram finder on the fellowship of the ring
```

```python
# scoring the bigram finder for the bible and the fellowship of the ring
bible_scored = biblefinder.score_ngrams(bigram_measures.raw_freq)
fellowship_scored = fellowshipfinder.score_ngrams(bigram_measures.raw_freq)
```

```python
# printing the top 50 scores from the bible
for bscore in bible_scored[:50]:
    print('The Bible Bigram Scores: ', bscore)

# printing the top 50 scores from the fellowship of the ring
for cscore in fellowship_scored[:50]:
    print('The Fellowship of the Ring Bigram Scores: ', cscore)
```

```python
# we can run bigram scores with a filter for alphs and stopwords but we already have that done so we will just create a new finder

biblefinder2 = BigramCollocationFinder.from_words(biblewords2) # running the bigram finder on the bible
fellowshipfinder2 = BigramCollocationFinder.from_words(fellowshipwords2) # running the bigram finder on the fellowship of the ring

# apply a filter that removes words that occur under a minimum threshold
biblefinder2.apply_freq_filter(2)
fellowshipfinder2.apply_freq_filter(2)

# we dont need to filter out words based on length because most the bigrams have words over 3 letters anyway

# scoring the second bigram finder for the bible and the fellowship of the ring
bible_scored2 = biblefinder2.score_ngrams(bigram_measures.raw_freq)
fellowship_scored2 = fellowshipfinder2.score_ngrams(bigram_measures.raw_freq)

# printing the top 20 scores from the second bible finder
for bscore in bible_scored2[:50]:
    print('The Bible Bigram Scores without stopwords: ', bscore)

# printing the top 20 scores from the second fellowship of the ring finder
for cscore in fellowship_scored2[:50]:
    print('The Fellowship of the Ring Bigram Scores without stopwords: ', cscore)
```

```python
# creating the PMI scores
# I used the wordslists that already have the alpha and stopwords filters on them so that I am getting the scores of the filtered pairs
biblefinder3 = BigramCollocationFinder.from_words(biblewords2) # running the bigram finder on the bible
fellowshipfinder3 = BigramCollocationFinder.from_words(fellowshipwords2) # running the bigram finder on the fellowship of the ring

# applying frequency filter of 5
biblefinder3.apply_freq_filter(5)
fellowshipfinder3.apply_freq_filter(5)

# scoring the third bigram finder for the bible and the fellowship of the ring
bible_scored3 = biblefinder3.score_ngrams(bigram_measures.pmi)
fellowship_scored3 = fellowshipfinder3.score_ngrams(bigram_measures.pmi)

# printing the top 50 scores from the second bible finder
for bscore in bible_scored3[:50]:
    print('The Bible PMI Scores: ', bscore)

# printing the top 50 scores from the second fellowship of the ring finder
for cscore in fellowship_scored3[:50]:
    print('The Fellowship of the Ring PMI Scores: ', cscore)
```
